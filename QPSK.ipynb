# Importing the libraries

```python
import numpy as np
from scipy import signal
import pandas as pd
import matplotlib.pyplot as plt
```

# QPSK

## Parameters for carrier function

```python
fc = int(input("Enter frequency of carrier signal: "))
fm = int(input("Enter frequency of modulating signal: "))
A = int(input("Enter amplitude of carrier signal: ")) 
i = int(input("Enter number of inputs: "))
```

## Sampling rate

```python
t = np.linspace(start=0, stop=1, num=10000, endpoint=True)
```

## Carrier signal

```python
carrier0 = np.array(A*np.sin(2*np.pi*fc*t + (2*0+1)*(np.pi / 4))).T
carrier1 = np.array(A*np.sin(2*np.pi*fc*t + (2*1+1)*(np.pi / 4))).T
carrier2 = np.array(A*np.sin(2*np.pi*fc*t + (2*2+1)*(np.pi / 4))).T
carrier3 = np.array(A*np.sin(2*np.pi*fc*t + (2*3+1)*(np.pi / 4))).T
```

## Plotting the output

```python
fig, axes = plt.subplots(nrows=4, ncols=1)
axes[0].plot(t, carrier0)
axes[1].plot(t, carrier1)
axes[2].plot(t, carrier2)
axes[3].plot(t, carrier3)

axes[0].set_title('Carrier wave for i = 0')
axes[1].set_title('Carrier wave for i = 1')
axes[2].set_title('Carrier wave for i = 2')
axes[3].set_title('Carrier wave for i = 3')

plt.tight_layout()

fig.set_figheight(12)
fig.set_figwidth(20)

axes[0].set_xlabel('time')
axes[0].set_ylabel('Amplitude')
axes[1].set_xlabel('time')
axes[1].set_ylabel('Amplitude')
axes[2].set_xlabel('time')
axes[2].set_ylabel('Amplitude')
axes[3].set_xlabel('time')
axes[3].set_ylabel('Amplitude')

axes[0].grid(which='major', axis='both')
axes[1].grid(which='major', axis='both')
axes[2].grid(which='major', axis='both')
axes[3].grid(which='major', axis='both')
```
